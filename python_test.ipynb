# function to calculate the age till now:

```python
from datetime import datetime as dt 

dob = input('enter your date of birth to calculate the age: ')
def agecaluclator(dob: str) -> str:
    dob = dt.strptime(dob , "%Y-%m-%d")
    now = dt.now()
    age = now - dob
    return age
print(agecaluclator(dob))
```

# create a function to check age eligibility for given customer based on dob, function will take two input dob and eligibility age

```python
from datetime import datetime

def check_eligibility(dob_str, eligibility_age):
    dob = datetime.strptime(dob_str, "%Y-%m-%d")
    today = datetime.now()

    age = today.year - dob.year
    if (today.month, today.day) < (dob.month, dob.day):
        age -= 1

    if age >= eligibility_age:
        return "Eligible"
    else:
        return "Not Eligible"


dob = input("Enter DOB (YYYY-MM-DD): ")
eligibility_age = int(input("Enter eligibility age: "))

print(check_eligibility(dob, eligibility_age))
```

# calculate total hours based on input sign in and sign out time

```python
from datetime import datetime

def calculateworkhours(sign_in, sign_out):
    in_time = datetime.strptime(sign_in, "%H:%M")
    out_time = datetime.strptime(sign_out, "%H:%M")
    duration = out_time - in_time
    total_hours = duration.total_seconds() / 3600
    return total_hours


sign_in = input("Enter sign-in time (HH:MM): ")
sign_out = input("Enter sign-out time (HH:MM): ")

print("Total working hours:", calculateworkhours(sign_in, sign_out))
```
