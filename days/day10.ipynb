```python
def our_decorator(func):
    def function_wrapper(x):
        c = 90 + 30
        print(str(c)+ ' Before Callling: '+func.__name__)
        func(x)
        d = 67 + 57
        print(c , d)
        print(c, d , ' After Calling: '+func.__name__)
    return function_wrapper

@our_decorator
def foo(x):
    print(' Hi Foo has been called with: '+str(x))

foo(100)
    
```

```python
def my_decorator(func): # my_decorator(greet('sahil'))
    def wrapper(name): # wrapper('sahil')
        print(name) # print('sahil')
        print("---- Before Calling Function ----")
        func(name) # greet('sahil')
        print("---- After Calling Function ----")
    return wrapper

@my_decorator
def greet(name1):
    print("Hello", name1)

greet("Sahil")
```

```python
def argument_test_natural_number(f):
    def helper(x):
        if type(x) == int and x > 0:
            print('GOOD')
            return f(x)
        else:
            print('Argument is not an integer and must be greater then 0')
    return helper

@argument_test_natural_number
def even_odd(num):
    if num % 2 == 0 :
        print('even number')
    else: 
        print('odd number')

even_odd('sahil')
print('------next value: ---------')
even_odd(8)
```

```python
class SahilExample:

    company = "Sahil's Company"  

    def __init__(self, name, salary):
        self.name = name        
        self.salary = salary

    def show_employee(self):
        print("Employee Name:", self.name)
        print("Salary:", self.salary)
        print("Company:", SahilExample.company)

    @classmethod
    def change_company(cls, new_name):
        cls.company = new_name
        print("Company name changed to:", cls.company)

    @staticmethod
    def is_high_salary(amount):
        if amount > 50000:
            return "High Salary"
        else:
            return "Low Salary"


emp1 = SahilExample("Sahil", 60000)

emp1.show_employee()

SahilExample.change_company("tata company")

emp1.show_employee()

print('------- static function : on class ')
result = SahilExample.is_high_salary(40000)
print(result)
print('--------------- emp1:sahil object salary (60,000) ------------')
print(emp1.salary)
print('-----------changed using static method---------------')
#emp1.is_high_salary()
result = emp1.is_high_salary(emp1.salary)
print(result)
print('--------------- emp1:sahil object salary (60,000) ------------')
print(emp1.salary)
```
