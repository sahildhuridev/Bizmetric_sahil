```python
def sub2(d):
    tot = 0
    for i in d:
        tot = tot + i
        print(i,tot)
        return tot
```

```python
def sub3(d):
    tot = 0
    for i in d:
        tot += i
        print(i,tot)
        print(f'{tot} first time')
        yield 1
        print(f'{tot} second time')
        yield 2
        print(f'{tot} third time')
        yield 3
```

```python
abc = sub3(d)
```

```python
c_name = ['mumbai', 'pune', 'banglore']

def city_generator_y(c_name):
    for i in c_name:
        yield i

cgy = city_generator_y(c_name)

while True:
    try:
        print(next(cgy))
    except StopIteration:
        print("Reached end, reversing list")
        c_name.reverse()
        cgy = city_generator_y(c_name)
        break
        
```

```python
# fibonacci series
def fibonacci(max1):
    a , b = 0 ,1 
    while a < max1:
        yield a 
        a, b = b , a+b
```

```python
obj=fibonacci(10)
```

```python
next(obj)
```

```python
import os
import uuid
from datetime import datetime


class SahilRestaurantSystem:

    # Predefined Menu (Item : Price)
    menu = {
        1: ("Burger", 120),
        2: ("Pizza", 250),
        3: ("Pasta", 180),
        4: ("Coke", 40),
        5: ("Fries", 90)
    }

    def __init__(self):
        self.order_dict = {}
        self.table_number = None
        self.order_id = str(uuid.uuid4())[:8]   

    def take_order(self):
        print("\n------ MENU ------")
        for key, value in self.menu.items():
            print(f"{key}. {value[0]} - ₹{value[1]}")

        self.table_number = input("\nEnter Table Number: ")

        while True:
            try:
                item_choice = int(input("\nEnter item number (0 to stop): "))

                if item_choice == 0:
                    break

                if item_choice not in self.menu:
                    print("Invalid item number.")
                    continue

                quantity = int(input("Enter quantity: "))

                if quantity <= 0:
                    print("Quantity must be positive.")
                    continue

                item_name = self.menu[item_choice][0]
                item_price = self.menu[item_choice][1]

                self.order_dict[item_name] = {
                    "price": item_price,
                    "quantity": quantity
                }

            except ValueError:
                print("Please enter valid numeric input.")


    def deliver_the_order(self):
        print("\nPreparing your order...")
        print("Order delivered successfully to Table", self.table_number)


    def generate_bill(self):
        total = 0
        bill_lines = []

        bill_lines.append("-" * 50)
        bill_lines.append(f"| {'SAHIL RESTAURANT BILL':^46} |")
        bill_lines.append("-" * 50)
        bill_lines.append(f"| Table No : {self.table_number:<34} |")
        bill_lines.append(f"| Order ID : {self.order_id:<34} |")
        bill_lines.append("-" * 50)
        bill_lines.append(f"| {'Item':<15} | {'Qty':<5} | {'Price':<7} | {'Total':<8} |")
        bill_lines.append("-" * 50)

        for item, details in self.order_dict.items():
            qty = details["quantity"]
            price = details["price"]
            item_total = qty * price
            total += item_total

            bill_lines.append(f"| {item:<15} | {qty:<5} | {price:<7} | {item_total:<8} |")

        bill_lines.append("-" * 50)
        bill_lines.append(f"| {'Grand Total':<31} ₹{total:<10} |")
        bill_lines.append("-" * 50)

        self.final_bill = "\n".join(bill_lines)

    def print_bill(self):
        filename = f"{self.table_number}_{self.order_id}.txt"

        with open(filename, "w") as file:
            file.write(self.final_bill)

        print("\nBill Generated Successfully!")
        print("Saved as:", filename)
        print("\n", self.final_bill)


if __name__ == "__main__":
    sahil_system = SahilRestaurantSystem()

    sahil_system.take_order()
    sahil_system.deliver_the_order()
    sahil_system.generate_bill()
    sahil_system.print_bill()
```
